```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python

df = pd.read_csv('/content/2015.csv')
print(df.head())
```

```python
df.info()
```

```python
df.describe()
```

```python

print("First 5 rows:\n", df.head())
print("\nDataset Info:\n")

print("\nMissing Values:\n", df.isnull().sum())

df.rename(columns={
    'Happiness Score': 'Score',
    'Economy (GDP per Capita)': 'GDP',
    'Health (Life Expectancy)': 'Health',
    'Trust (Government Corruption)': 'Trust'
}, inplace=True)

top10 = df.nlargest(10, 'Score')
bottom10 = df.nsmallest(10, 'Score')

print("\nTop 10 Happiest Countries:\n", top10[['Country', 'Score']])
print("\nBottom 10 Countries:\n", bottom10[['Country', 'Score']])


plt.figure(figsize=(8,5))
sns.barplot(x='Score', y='Country', data=top10)
plt.title("Top 10 Happiest Countries")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,5))
sns.barplot(x='Score', y='Country', data=bottom10)
plt.title("Bottom 10 Countries")
plt.tight_layout()
plt.show()

plt.figure()
sns.scatterplot(x='Freedom', y='Score', data=df)
plt.title("Freedom vs Happiness Score")
plt.show()

region_avg = df.groupby('Region')['Score'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,5))
region_avg.plot(kind='bar')
plt.title("Average Happiness Score by Region")
plt.ylabel("Happiness Score")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

print("\nKey Insights:")
print("- Higher GDP generally increases happiness.")
print("- Health and life expectancy strongly affect happiness.")
```
